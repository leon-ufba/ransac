```python
def FPGAout_read(filename):
    data = []
    columns = ['step', 'model', 'bestfit', 'inliers', 'angle', 'distance']

    with open(filename, 'r') as file:
        while True:
            result = {}
            test_case = file.readline().strip()
            if not test_case:
                break
            result['step'] = int(test_case)
            result['model'] = list(map(float, file.readline()[1:-2].split(',')))
            result['bestfit'] = float(file.readline().strip())
            result['inliers'] = int(file.readline().strip())
            result['angle'] = float(file.readline().strip())
            result['distance'] = float(file.readline().strip())
            file.readline()
            data.append(result)

    return pd.DataFrame(data, columns=columns)
```

```python
FPGA_out_PC = FPGAout_read(lista_arquivos[1])
FPGA_out_NIOS = FPGAout_read(lista_arquivos[3])
FPGA_out_NIOS_H = FPGAout_read(lista_arquivos[2])
```

```python
modelo_df1 = pd.DataFrame(FPGA_out_PC['model'].tolist(), columns=['python_angular', 'python_linear'])
modelo_df2 = pd.DataFrame(FPGA_out_NIOS['model'].tolist(), columns=['nios_angular', 'nios_linear'])

# Concatenar as colunas "modelo" dos dois DataFrames em um novo DataFrame
modelo_concat = pd.concat([modelo_df1, modelo_df2], axis=1)

modelo_concat['erro_angular'] = abs(modelo_concat['python_angular'] - modelo_concat['nios_angular'])
modelo_concat['erro_linear'] = abs(modelo_concat['python_linear'] - modelo_concat['nios_linear']) 
```
